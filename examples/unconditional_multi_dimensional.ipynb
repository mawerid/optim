## Himmelblau Function

The Himmelblau function is a multi-modal function used to test optimization algorithms. It is defined as:

$\ f(x, y) = (x^2 + y - 11)^2 + (x + y^2 - 7)^2 $

### Extremum Points

The function has one local maximum and four local minima:

- **Local Maximum:**
  - $\ f(-0.270845, -0.923039) = 181.617 $

- **Local Minima:**
  - $\ f(3.0, 2.0) = 0.0 $
  - $\ f(-2.805118, 3.131312) = 0.0 $
  - $\ f(-3.779310, -3.283186) = 0.0 $
  - $\ f(3.584428, -1.848126) = 0.0 $

```python
func = lambda point: (point[0] ** 2 + point[1] - 11) ** 2 + (point[0] + point[1] ** 2 - 7) ** 2
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = func(np.array([x, y]))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

## Rosenbrock Function

The Rosenbrock function, also known as the Rosenbrock's valley or Rosenbrock's banana function, is a non-convex function used to test optimization algorithms. It is defined as:

$\ f(x, y) = (1 - x)^2 + 100(y - x^2)^2 $

### Extremum Points

The function has one global minimum:

- **Global Minimum:**
  - $\ f(1, 1) = 0 $


```python
func = lambda point: (1 - point[0]) ** 2 + 100 * (point[1] - point[0] ** 2) ** 2
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z = func(np.array([x, y]))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```
